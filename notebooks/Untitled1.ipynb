```python
import rowgenerators as rg

pop = rg.dataframe('census:/2016/5/CA/tract/B01003')
geo = pop.geoframe()
```

```python
df = geo.set_index('geoid').join(pop, rsuffix='_r')
```

```python
df['density'] = df.b01003_001 / df.aland * 1_000_000
```

```python
df[['density','geometry']].to_csv('/tmp/density.csv')
```
