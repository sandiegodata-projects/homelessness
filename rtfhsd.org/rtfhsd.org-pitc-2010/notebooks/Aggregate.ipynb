```python
pitc = doc.resource('pitc')\
    .geoframe()\
    .groupby(['tract','observe'])\
    .geometry.count().to_frame()\
    .unstack(-1)\
    .fillna(0)

pitc.columns = ['handbuilt','individual','vehicle']
pitc['total'] = pitc.sum(axis=1)

pitc
```
